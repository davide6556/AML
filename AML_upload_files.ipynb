## Load CSV Files with the Python Standard Library

```python
import csv
import numpy as np
filename = 'pima-indians-diabetes.data.csv'
raw_data = open(filename, 'r')
reader = csv.reader(raw_data, delimiter=',', quoting=csv.QUOTE_NONE)
x = list(reader) # We have a list of list in the dataset
print(type(x))
```

```python
x[0:10]
```

```python
print(x[0])
```

```python
# we want to have an array, so we tranform the list in this way
mydata_withpython = np.array(x).astype('float')
print(type(mydata_withpython))
```

```python
print(mydata_withpython.shape)
# 9 columns and 768 rows
# 1 col will be the column with the label 0/1
```

## Load CSV Files with NumPy

```python
from numpy import loadtxt
filename = 'pima-indians-diabetes.data.csv' # specify the data we want
raw_data = open(filename, 'rt') # we open the file in the mod read text
mydata_withnumpy = loadtxt(raw_data, delimiter=",")
```

```python
mydata_withnumpy
```

```python
print(mydata_withnumpy.shape)
```

## Load CSV Files with Pandas

```python
from pandas import read_csv
filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
mydata_withpandas = read_csv(filename, names=names)
# The function returns a `pandas.DataFrame` that one can immediately process, summarize, plot, etc
# better for visualization
print(type(mydata_withpandas))
```

```python
print(mydata_withpandas.shape)
```

```python
mydata_withpandas
```
